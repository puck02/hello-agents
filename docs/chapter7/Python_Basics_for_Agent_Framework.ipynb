### 知识点

- **字符串方法**：字符串后面加 `.方法名()` 可以做各种操作，比如 `.lower()` 转小写、`.strip()` 去空格。

````python
# === 1.1 变量：给数据起名字 ===
name = "AI助手"        # 字符串
age = 3                # 数字
is_smart = True        # 布尔值（对/错）

print(name)   # → AI助手
print(age)    # → 3

# === 1.2 f-string：在字符串里嵌入变量 ===
# 引号前加一个 f，然后用 {变量名} 插入值
greeting = f"你好！我是 {name}，今年 {age} 岁了。"
print(greeting)

# 在第七章中，Agent 初始化完成后会这样打印：
# print(f"✅ {name} 初始化完成，工具调用: {'启用' if is_smart else '禁用'}")
#       ↑ 插入变量              ↑ {} 里面还可以写简单的判断逻辑
print(f"✅ {name} 初始化完成，智能模式: {'启用' if is_smart else '禁用'}")

# === 1.3 字符串常用方法 ===
# 第七章用这些方法来处理 URL 和 API 密钥
url = "  https://API.OpenAI.COM/v1  "
print(url.strip())         # 去掉两边空格 → "https://API.OpenAI.COM/v1"
print(url.strip().lower()) # 全部变小写   → "https://api.openai.com/v1"

text = "action=search,query=Python"
print(text.split(","))     # 按逗号分割 → ['action=search', 'query=Python']

key = "ms-abcdefg"
print(key.startswith("ms-"))  # 是否以 "ms-" 开头 → True
# 第七章用 startswith 来判断密钥属于哪个服务商
````

## 2. 列表和字典 📦

### 在第七章哪里用到了？
```python
messages = []                                  # 创建空列表
messages.append({"role": "user", "content": "你好"})  # 往列表里加字典
```

### 知识点
- **列表 (list)**：用 `[]` 创建，可以装多个东西，按位置取值（从 0 开始数）。
- **字典 (dict)**：用 `{}` 创建，里面是 `"键": 值` 的配对。像电话簿一样，通过名字查号码。
- **`.get(键, 默认值)`**：安全地从字典取值。找不到不会报错，而是返回默认值。
- **切片 `[:3]`**：从列表截取前 3 个元素。不够 3 个也不会报错。
- **`enumerate(列表, 起始号)`**：遍历时自动编号，从指定数字开始。

````python
# === 2.1 列表：有序的数据容器 ===
fruits = ["苹果", "香蕉", "橙子"]
print(fruits[0])   # 取第1个（从0开始）→ 苹果
print(fruits[-1])  # 取最后1个 → 橙子

fruits.append("葡萄")  # 往末尾添加
print(fruits)          # → ['苹果', '香蕉', '橙子', '葡萄']
print(len(fruits))     # 长度 → 4

# 第七章中，messages 就是一个列表，里面装着字典
messages = []
messages.append({"role": "system", "content": "你是AI助手"})
messages.append({"role": "user", "content": "你好"})
print(f"消息列表: {messages}")

# === 2.2 字典：通过名字查数据 ===
agent_info = {"name": "小助手", "provider": "openai", "temperature": 0.7}
print(agent_info["name"])  # → 小助手

# ⚠️ 危险写法：键不存在就崩溃
try:
    print(agent_info["model"])
except KeyError:
    print("❌ 字典里没有 'model' 这个键，报错了！")

# ✅ 安全写法：.get() — 第七章到处都在用
model = agent_info.get("model", "gpt-3.5-turbo")  # 没有就用默认值
print(f"模型: {model}")  # → gpt-3.5-turbo

# === 2.3 列表切片 ===
numbers = [10, 20, 30, 40, 50]
print(numbers[:3])   # 前3个 → [10, 20, 30]
print(numbers[1:4])  # 第2到第4个 → [20, 30, 40]
# 第七章搜索结果只取前3条：response.get('results', [])[:3]

# === 2.4 enumerate：带序号遍历 ===
tools = ["search", "calculator", "weather"]
for i, tool in enumerate(tools, 1):  # 从1开始编号
    print(f"工具 {i}: {tool}")
# 工具 1: search
# 工具 2: calculator
# 工具 3: weather
````

## 3. 条件判断、循环与异常处理 🚦

### 在第七章哪里用到了？
```python
# if/elif/else 判断（自动检测 provider）
if provider == "modelscope":
    ...
elif provider == "openai":
    ...
else:
    super().__init__(...)

# while 循环（ReAct Agent 核心）
while current_step < self.max_steps:
    current_step += 1
    ...
    if action.startswith("Finish"):
        break

# try/except 异常处理（防止崩溃）
try:
    return self._search_tavily(query)
except Exception as e:
    print(f"⚠️ 搜索失败: {e}")
```

### 知识点
- **`if / elif / else`**：根据条件选择执行哪段代码。
- **`==`**：判断是否相等（两个等号！一个等号 `=` 是赋值）。
- **`and` / `or` / `not`**：组合多个条件。
- **`while`**：只要条件为 True，就反复执行。`break` 可以提前跳出。
- **`try / except`**：把"可能出错"的代码放在 `try` 里，出错后跳到 `except`，程序不会崩溃。

````python
# === 3.1 if / elif / else ===
provider = "modelscope"

if provider == "modelscope":
    print("使用 ModelScope 平台")
elif provider == "openai":
    print("使用 OpenAI 平台")
elif provider == "ollama":
    print("使用本地 Ollama 服务")
else:
    print("使用默认配置")

# and 组合条件（第七章里判断是否启用工具）
tool_registry = None
enable_tool_calling = True
# 两个条件都满足才为 True
result = enable_tool_calling and tool_registry is not None
print(f"是否启用工具: {result}")  # False，因为 tool_registry 是 None

# === 3.2 while 循环（ReAct Agent 的核心逻辑） ===
max_steps = 3
current_step = 0

while current_step < max_steps:
    current_step += 1  # 等于 current_step = current_step + 1
    print(f"--- 第 {current_step} 步 ---")
    
    if current_step == 2:
        print("找到答案了！提前结束")
        break  # 跳出整个 while 循环

# === 3.3 try / except（防崩溃） ===
def divide(a, b):
    try:
        result = a / b
        print(f"{a} ÷ {b} = {result}")
    except ZeroDivisionError:
        print("❌ 不能除以零！")
    except Exception as e:
        # Exception 能接住任何错误
        print(f"❌ 出错了: {e}")

divide(10, 3)   # 正常
divide(10, 0)   # 被 except 接住，不会崩溃
````

### 知识点
- **`def`**：定义一个函数（可以反复使用的代码块）。
- **`return`**：把结果返回给调用者。执行到 `return` 后，函数立即结束。
- **默认参数**：`max_steps=5` 表示不传就用 5。
- **`"""文档字符串"""`**：写在函数开头的说明，告诉别人这个函数干什么。
- **`.format()`**：把字符串里的 `{占位符}` 替换成真实值。和 f-string 类似，但更适合处理预定义的模板。

````python
# === 4.1 定义和调用函数 ===
def greet(name):
    """向某人打招呼"""  # 这就是文档字符串
    return f"你好，{name}！"

result = greet("小明")
print(result)  # → 你好，小明！

# === 4.2 默认参数 ===
# 第七章中 max_steps 默认是 5，用户不传就用默认值
def search(query, max_results=3):
    print(f"搜索: {query}，最多返回 {max_results} 条")

search("Python")           # max_results 用默认值 3
search("Python", 10)       # 手动指定为 10

# === 4.3 .format() 模板替换 ===
# 第七章 ReAct Agent 的提示词模板就是用 .format() 填充的
template = """你是一个AI助手。
可用工具: {tools}
问题: {question}
"""

filled = template.format(
    tools="search, calculator",
    question="今天天气怎么样？"
)
print(filled)

# === 4.4 return 可以提前结束函数 ===
def check_input(text):
    if not text.strip():       # .strip() 去空格后如果是空的
        return "❌ 输入不能为空"  # 直接返回，后面的代码不会执行
    return f"✅ 收到: {text}"

print(check_input(""))       # → ❌ 输入不能为空
print(check_input("你好"))   # → ✅ 收到: 你好
````

## 6. 类和对象入门 🏠

### 在第七章哪里用到了？
```python
class Agent(ABC):
    def __init__(self, name, llm, system_prompt=None, config=None):
        self.name = name           # 保存属性
        self.llm = llm
        self._history = []         # _ 开头表示"内部使用"

    def add_message(self, message):
        self._history.append(message)

    def __str__(self):
        return f"Agent(name={self.name})"
```

### 知识点
想象一下**盖房子**：
- **类 (class)** = **图纸**。描述"这种东西有什么属性、能做什么"。
- **对象 (object)** = 根据图纸盖出来的**真实的房子**。

关键概念：
- **`__init__`**：构造函数。创建对象时**自动执行**，用来设置初始属性。
- **`self`**：代表**当前这个对象自己**。类里每个方法的第一个参数都是 `self`，但调用时不需要手动传。
- **`self.xxx = xxx`**：把数据保存到对象身上，这样对象的其他方法也能访问。
- **`_xxx`（单下划线开头）**：暗示"这是内部使用的，外部最好别直接碰"。
- **`__str__`**：当你 `print(对象)` 时，Python 会自动调用这个方法。

````python
# === 6.1 定义一个简单的类 ===
class Robot:
    def __init__(self, name, color):
        # self = 正在创建的这个机器人自己
        self.name = name       # 把名字保存到自己身上
        self.color = color
        self._battery = 100    # _ 开头 = 内部属性
        print(f"🤖 制造了 {self.name}")

    def say_hello(self):
        print(f"你好！我是 {self.name}，{self.color}的。电量: {self._battery}%")

    def __str__(self):
        return f"Robot(name={self.name}, color={self.color})"

# === 6.2 创建对象（实例化） ===
bot1 = Robot("小白", "白色")
bot2 = Robot("小黑", "黑色")

# 调用方法时不需要传 self，Python 自动处理
bot1.say_hello()
bot2.say_hello()

# 直接 print 对象，会调用 __str__
print(bot1)  # → Robot(name=小白, color=白色)

# === 6.3 对比第七章的 Agent 类 ===
# Agent.__init__ 做了类似的事：
#   self.name = name          保存名字
#   self.llm = llm            保存 LLM 实例
#   self.system_prompt = ...  保存系统提示词
#   self._history = []        创建空的历史记录列表
````

## 7. 继承与 super() 🏗️

### 在第七章哪里用到了？
```python
# 框架提供了父类
class Agent(ABC):
    def __init__(self, name, llm, system_prompt=None, config=None):
        self.name = name
        self.llm = llm
        ...

# 你写的代码继承父类，加上自己的功能
class MySimpleAgent(SimpleAgent):     # ← 继承
    def __init__(self, name, llm, system_prompt=None, ...):
        super().__init__(name, llm, system_prompt, config)  # ← 先调用父类
        self.tool_registry = tool_registry  # 再加自己的东西
```

### 知识点
- **继承**：`class 子类(父类)` — 子类自动拥有父类的所有属性和方法。就像孩子继承父母的基因。
- **`super()`**：代表父类。`super().__init__(...)` = "让父类先把它该做的初始化做完"。
- **为什么必须 super()?** 不调用的话，父类 `__init__` 里定义的属性（如 `self.name`）不存在，后续使用就会报错。
- **重写 (Override)**：子类可以写一个和父类同名的方法，调用时会优先用子类的版本。

````python
# === 7.1 父类（框架提供的） ===
class Animal:
    def __init__(self, name, age):
        self.name = name
        self.age = age
        print(f"🐾 动物 {self.name} 登记完毕，年龄 {self.age}")

    def speak(self):
        print(f"{self.name} 发出了声音")

# === 7.2 子类：继承 Animal，加上自己的特性 ===
class Dog(Animal):
    def __init__(self, name, age, breed):
        # 关键！先调用父类的 __init__，让 self.name 和 self.age 被创建
        super().__init__(name, age)
        # 再设置自己特有的属性
        self.breed = breed
        print(f"🐕 品种: {self.breed}")

    # 重写父类的 speak 方法
    def speak(self):
        print(f"{self.name} 说: 汪汪汪！")

    # 子类自己的新方法
    def fetch(self):
        print(f"{self.name} 去捡球了！")

dog = Dog("旺财", 3, "金毛")
dog.speak()  # 调用的是子类重写的版本
dog.fetch()  # 子类自己的方法

# === 7.3 如果忘记调用 super() 会怎样？ ===
class BadDog(Animal):
    def __init__(self, name, age, breed):
        # 故意不调用 super().__init__()
        self.breed = breed

try:
    bad = BadDog("来福", 2, "哈士奇")
    bad.speak()  # ❌ 报错！因为 self.name 从未被创建
except AttributeError as e:
    print(f"\n❌ 报错了：{e}")
    print("   原因：没调用 super().__init__()，所以 self.name 不存在")
````

## 8. 抽象基类 ABC 📜

### 在第七章哪里用到了？
```python
from abc import ABC, abstractmethod

class Agent(ABC):
    """Agent基类"""

    @abstractmethod
    def run(self, input_text: str, **kwargs) -> str:
        """所有子类必须实现这个方法"""
        pass
```

### 知识点
- **ABC (Abstract Base Class)**：一种特殊的类，**不能直接创建对象**，只能被继承。相当于一份"强制合同"。
- **`@abstractmethod`**：标记某个方法为"抽象的"——我只声明了名字，具体怎么做你在子类里自己写。
- **强制实现**：如果子类没有实现所有 `@abstractmethod` 标记的方法，Python 会直接报 `TypeError`。
- **为什么这样设计？** 框架保证不管谁写的 Agent，都有 `run()` 方法，用户只需统一调用 `agent.run("问题")`。

````python
from abc import ABC, abstractmethod

# === 8.1 定义抽象基类 ===
class BaseAgent(ABC):
    def __init__(self, name):
        self.name = name

    @abstractmethod
    def run(self, input_text):
        """每个 Agent 必须实现这个方法"""
        pass

    # 普通方法不需要 @abstractmethod，子类可以直接继承
    def get_name(self):
        return self.name

# === 8.2 抽象类不能直接创建对象 ===
try:
    agent = BaseAgent("测试")
except TypeError as e:
    print(f"❌ {e}")

# === 8.3 不实现抽象方法的子类也不行 ===
try:
    class LazyAgent(BaseAgent):
        pass  # 没有实现 run()

    lazy = LazyAgent("懒惰")
except TypeError as e:
    print(f"❌ {e}")

# === 8.4 正确实现 ===
class SmartAgent(BaseAgent):
    def run(self, input_text):
        return f"🤖 {self.name} 处理了: {input_text}"

agent = SmartAgent("智能助手")
result = agent.run("写一首诗")
print(f"✅ {result}")
print(f"✅ 名字: {agent.get_name()}")  # 继承来的方法可以直接用
````

```python
def run(self, input_text: str, **kwargs) -> str:     # 参数和返回值类型
system_prompt: Optional[str] = None                   # 可选参数
current_history: List[str] = []                       # 字符串列表
metadata: Optional[Dict[str, Any]] = None             # 可选字典
MessageRole = Literal["user", "assistant", "system"]  # 限制取值
```

### 知识点
**类型提示不会报错！** Python 不会因为你写了 `age: int` 就阻止你传字符串——它只是一种"说明书"，帮你和别人理解代码。

| 写法 | 含义 | 例子 |
|------|------|------|
| `str` | 字符串 | `name: str` |
| `int` | 整数 | `age: int` |
| `float` | 小数 | `temperature: float` |
| `bool` | 布尔值 | `debug: bool` |
| `Optional[str]` | 字符串**或** None | 可传可不传的参数 |
| `List[str]` | 字符串列表 | `["a", "b"]` |
| `Dict[str, Any]` | 键是字符串，值随意 | `{"key": 任意}` |
| `Any` | 任何类型 | 不限制 |
| `Literal["a","b"]` | 只能是 "a" 或 "b" | 限制取值范围 |
| `-> str` | 函数返回字符串 | 写在函数定义后面 |

````python
from typing import Optional, List, Dict, Any, Literal

# === 9.1 基本类型提示 ===
name: str = "AI助手"
age: int = 3
temperature: float = 0.7
is_active: bool = True

# === 9.2 Optional：可以是值，也可以是 None ===
# 第七章中 system_prompt 是可选的
def create_agent(name: str, prompt: Optional[str] = None):
    if prompt is None:
        print(f"{name}: 没有设置提示词，使用默认值")
    else:
        print(f"{name}: 提示词是 '{prompt}'")

create_agent("助手A")                     # 不传 prompt → None
create_agent("助手B", "你是一个翻译官")    # 传了 prompt

# === 9.3 List 和 Dict ===
tools: List[str] = ["search", "calculator"]
config: Dict[str, Any] = {
    "model": "gpt-4",
    "temperature": 0.7,
    "debug": False
}
print(f"工具: {tools}")
print(f"配置: {config}")

# === 9.4 Literal：限制取值 ===
# 第七章用它限制消息角色只能是这4种
MessageRole = Literal["user", "assistant", "system", "tool"]

def send_message(content: str, role: MessageRole):
    print(f"[{role}] {content}")

send_message("你好", "user")
send_message("你好！有什么可以帮你的？", "assistant")

# === 9.5 函数返回值类型 ===
def add(a: int, b: int) -> int:    # -> int 表示返回整数
    return a + b

def greet(name: str) -> str:       # -> str 表示返回字符串
    return f"你好，{name}"

print(add(1, 2))
print(greet("小明"))
````

## 10. Pydantic 数据校验 ✅

### 在第七章哪里用到了？
```python
from pydantic import BaseModel

class Message(BaseModel):
    content: str
    role: MessageRole
    timestamp: datetime = None
    metadata: Optional[Dict[str, Any]] = None

class Config(BaseModel):
    default_model: str = "gpt-3.5-turbo"
    temperature: float = 0.7
    debug: bool = False
```

### 知识点
- **Pydantic**：一个 Python 库，帮你定义"数据长什么样"，并**自动检查**。
- **继承 `BaseModel`**：告诉 Pydantic "这个类是一个数据模型，请帮我校验"。
- **自动校验**：传了错误类型 → 直接报错，防止脏数据进入系统。
- **自动转换**：能转的会帮你转（如字符串 "20" → 整数 20）。
- **默认值**：`temperature: float = 0.7` 不传就用 0.7。

和普通 class 的区别：**普通 class 不管你传什么都照单全收，Pydantic 会帮你把关。**

````python
from pydantic import BaseModel, ValidationError
from typing import Optional, Dict, Any, Literal
from datetime import datetime

# === 10.1 模仿第七章的 Message 类 ===
MessageRole = Literal["user", "assistant", "system", "tool"]

class Message(BaseModel):
    content: str
    role: MessageRole
    timestamp: datetime = None
    metadata: Optional[Dict[str, Any]] = None

# 创建一条消息
msg = Message(content="你好", role="user")
print(f"✅ 消息: [{msg.role}] {msg.content}")
print(f"   时间戳: {msg.timestamp}")  # 默认值 None

# === 10.2 校验功能 ===
try:
    # role 只能是 "user"/"assistant"/"system"/"tool"
    bad_msg = Message(content="测试", role="hacker")  # ❌ 不合法
except ValidationError as e:
    print(f"\n❌ 校验失败（role 不合法）:")
    print(f"   {e.errors()[0]['msg']}")

# === 10.3 模仿第七章的 Config 类 ===
class Config(BaseModel):
    default_model: str = "gpt-3.5-turbo"
    temperature: float = 0.7
    debug: bool = False

# 不传参数 → 全用默认值
config = Config()
print(f"\n✅ 默认: 模型={config.default_model}, 温度={config.temperature}")

# 只改你想改的
config2 = Config(temperature=0.9, debug=True)
print(f"✅ 自定义: 温度={config2.temperature}, 调试={config2.debug}")
````

## 11. 万能参数 `**kwargs` 🎒

### 在第七章哪里用到了？
```python
# Agent 基类
def run(self, input_text: str, **kwargs) -> str:
    pass

# MySimpleAgent 把 kwargs 原封不动传给 LLM
response = self.llm.invoke(messages, **kwargs)

# MyLLM 初始化时从 kwargs 取值
def __init__(self, model=None, ..., **kwargs):
    self.temperature = kwargs.get('temperature', 0.7)
```

### 知识点
- **`**kwargs`**：把所有"多出来的、带名字的参数"打包成一个**字典**。
- **`kwargs.get('key', 默认值)`**：从字典里取值（和普通字典用法一样）。
- **透传**：`**kwargs` 最大的用处——"我不管你传了什么参数，我原封不动帮你转交给下一个函数"。

流程示意：
1. 用户调用 `agent.run("问题", temperature=0.9)`
2. `run` 里收到 `kwargs = {"temperature": 0.9}`
3. `self.llm.invoke(messages, **kwargs)` 把整个字典展开传给 LLM
4. 这样 Agent 不需要知道 LLM 有哪些参数，传什么都能到达

````python
# === 11.1 **kwargs 基础用法 ===
def call_llm(prompt, **kwargs):
    print(f"提示词: {prompt}")
    print(f"额外参数: {kwargs}")  # kwargs 是一个字典
    
    temp = kwargs.get("temperature", 0.7)  # 取值，取不到用默认值
    print(f"温度: {temp}")

call_llm("你好")
print("-" * 30)
call_llm("写首诗", temperature=0.9, model="gpt-4")

# === 11.2 透传：原封不动传给下一个函数 ===
def inner_function(**kwargs):
    """模拟底层 LLM 调用"""
    print(f"  底层收到: {kwargs}")

def outer_function(user_input, **kwargs):
    """模拟 Agent 的 run 方法"""
    print(f"Agent 收到: {user_input}")
    # **kwargs 展开字典，变回关键字参数传给下一个函数
    inner_function(**kwargs)

print("-" * 30)
outer_function("你好世界", temperature=0.9, max_tokens=100)

# === 11.3 和普通参数混用 ===
# 第七章: def run(self, input_text: str, max_tool_iterations: int = 3, **kwargs)
def run(input_text, max_steps=3, **kwargs):
    print(f"输入: {input_text}")
    print(f"最大步数: {max_steps}")
    print(f"其他参数: {kwargs}")

print("-" * 30)
run("计算1+1", max_steps=5, temperature=0.8)
````

## 12. 正则表达式 🔍

### 在第七章哪里用到了？
```python
import re

# MySimpleAgent 用正则从 LLM 回复中提取工具调用指令
pattern = r'\[TOOL_CALL:([^:]+):([^\]]+)\]'
matches = re.findall(pattern, response)
```

### 知识点
正则表达式 = 一种"文本搜索模式"。告诉 Python "去这段文字里找符合这个规则的内容"。

**第七章用到的正则拆解：** `r'\[TOOL_CALL:([^:]+):([^\]]+)\]'`

| 部分 | 含义 |
|------|------|
| `r'...'` | 原始字符串，`\` 不会被转义 |
| `\[` | 匹配字面的 `[`（`[` 有特殊含义，需要 `\` 转义） |
| `TOOL_CALL:` | 必须完全匹配这个前缀 |
| `([^:]+)` | **捕获组1**：除冒号外的所有字符 → **工具名** |
| `:` | 冒号分隔符 |
| `([^\]]+)` | **捕获组2**：除 `]` 外的所有字符 → **参数** |
| `\]` | 匹配字面的 `]` |

`re.findall(pattern, text)` → 返回所有匹配到的**(捕获组1, 捕获组2)**组成的列表。

````python
import re

# === 12.1 简单的正则匹配 ===
text = "我的手机号是 13812345678，备用号码是 13998765432"
phones = re.findall(r'1\d{10}', text)  # 1开头 + 10个数字
print(f"找到的手机号: {phones}")

# === 12.2 第七章的核心正则 ===
# 模拟 LLM 返回的文字，里面夹带了工具调用指令
llm_response = """好的，让我帮你查一下。
[TOOL_CALL:weather:上海]
另外帮你算一下。
[TOOL_CALL:calculator:15 * 8 + 32]
最后搜索一些资料。
[TOOL_CALL:search:Python编程入门]
"""

pattern = r'\[TOOL_CALL:([^:]+):([^\]]+)\]'
# 拆解：
# \[TOOL_CALL:    → 匹配 [TOOL_CALL:
# ([^:]+)         → 捕获组1：除冒号外的字符 = 工具名
# :               → 冒号分隔符
# ([^\]]+)        → 捕获组2：除]外的字符 = 参数
# \]              → 匹配 ]

matches = re.findall(pattern, llm_response)

print(f"\n找到 {len(matches)} 个工具调用:")
for tool_name, params in matches:
    print(f"  🔧 工具: {tool_name}, 参数: {params}")

# findall 返回的原始数据
print(f"\n原始返回值: {matches}")
# → [('weather', '上海'), ('calculator', '15 * 8 + 32'), ('search', 'Python编程入门')]
````

## 13. 生成器与 yield 🌊

### 在第七章哪里用到了？
```python
# MySimpleAgent 的流式输出
def stream_run(self, input_text: str, **kwargs) -> Iterator[str]:
    for chunk in self.llm.stream_invoke(messages, **kwargs):
        full_response += chunk
        yield chunk   # 产出一点就送出去一点

# 使用方式
for chunk in agent.stream_run("什么是AI？"):
    print(chunk, end="")  # 像 ChatGPT 一样一个字一个字打出来
```

### 知识点
- **普通函数用 `return`**：算完了，一次性全部返回。函数结束。
- **生成器用 `yield`**：每次产出一个值，函数**暂停**，等下次再要时**继续执行**。
- **为什么用它？** 想象 ChatGPT 打字效果——字一个个蹦出来。LLM 返回文字是一点点的，用 `yield` 收到一点就送一点，用户不用干等。
- **怎么接收？** 用 `for` 循环遍历生成器即可。

````python
import time

# === 13.1 return vs yield ===

# return：一次性返回全部
def get_all_data():
    result = []
    for i in range(4):
        time.sleep(0.3)
        result.append(f"数据{i}")
    return result  # 全部攒好了再给

print("⏳ return 方式（要等很久）:")
print(get_all_data())

print("-" * 30)

# yield：来一个给一个
def stream_data():
    for i in range(4):
        time.sleep(0.3)
        yield f"数据{i}"  # 有一个就送一个
        print(f"  📦 送出了数据{i}，暂停...")

print("🌊 yield 方式（来一个收一个）:")
for item in stream_data():
    print(f"  👀 收到: {item}")

# === 13.2 模拟 ChatGPT 打字效果 ===
def fake_llm_stream(prompt):
    """模拟 LLM 的流式返回"""
    response = "人工智能是一门研究如何让计算机模拟人类智能的科学。"
    for char in response:
        time.sleep(0.05)
        yield char   # 一个字一个字送出

print("\n🤖 模拟 ChatGPT 打字效果:")
full_text = ""
for chunk in fake_llm_stream("什么是AI？"):
    print(chunk, end="", flush=True)  # end="" 不换行，flush 立即显示
    full_text += chunk
print(f"\n\n完整回复（{len(full_text)} 字）: {full_text}")
````
